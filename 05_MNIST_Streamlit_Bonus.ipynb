```python
# app.py
%%writefile streamlit_app.py
import streamlit as st
import tensorflow as tf
import numpy as np
from PIL import Image
import cv2

model = tf.keras.models.load_model('/content/AI_Tools_Assignment/data/mnist_cnn.h5')

st.title("âœï¸ Draw a Digit â†’ Get Prediction")
st.markdown("**99.1 % accurate CNN** â€¢ Draw below â€¢ Click **Predict**")

col1, col2 = st.columns(2)
with col1:
    canvas = st.canvas(fill_color="black", stroke_width=20, stroke_color="white",
                        background_color="black", height=280, width=280)
with col2:
    st.markdown("### Live Preview")
    if canvas.image_data is not None:
        st.image(canvas.image_data, width=280)

if st.button("ðŸ”® Predict", type="primary"):
    if canvas.image_data is not None:
        img = canvas.image_data.astype('uint8')
        img = cv2.resize(img, (28,28))
        img = cv2.cvtColor(img, cv2.COLOR_RGBA2GRAY)
        img = img.reshape(1,28,28,1).astype('float32')/255
        pred = model.predict(img, verbose=0)
        digit = np.argmax(pred)
        conf = pred.max()
        st.success(f"**Prediction: {digit}**")
        st.metric("Confidence", f"{conf:.1%}")
        st.bar_chart(pred.flatten())
    else:
        st.warning("Draw something first!")
```
