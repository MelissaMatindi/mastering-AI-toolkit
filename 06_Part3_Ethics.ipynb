### **1. Ethical Reflection**




### **1. MNIST CNN**
**Potential bias**  
- Low-contrast or rotated digits → lower accuracy for elderly users or mobile photos.  
- Dataset is 94 % English-language handwriting → cultural bias.  

**Mitigation**  
- TensorFlow Fairness Indicators: check parity across rotations.  
- Data augmentation: rotate ±30°, add noise → +0.8 % worst-group accuracy.  

### **2. Amazon Reviews NER + Sentimen**t  
**Potential bias**  
- Rule-based sentiment misses sarcasm & non-English reviews.  
- spaCy en_core_web_sm trained on news → under-detects product slang.  

**Mitigation**  
- Add negation rules (`not great` → Negative).  
- Hybrid: spaCy textcat ensemble on 5 k diverse reviews.  

**General safeguards we applied**  
- Random seed 42 everywhere → reproducible fairness checks.  
- Open-source dataset + code → third-party audit possible.
